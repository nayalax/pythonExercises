```python
import pandas as pd
import missingno as msno
import matplotlib.pyplot as plt

filePath = "../data/avgIQpercountry.csv"

df = pd.read_csv(filePath)
print(df.head())

print('------')
print(df.columns)
print('#######')
# Hay algunas columnas con espacios en blanco antres y después
# Esto nos puede dificultar el acceso, vamos a limpiar
df.columns = [column.strip() for column in df.columns]
# Ahora debería salir correctamente
print(df.columns)

# Identificar valores NaN
missing_values = df.isna().sum()
print(missing_values)

missing_gni = missing_values['GNI - 2021']

for index, item in missing_values.items():
    if item > 0:
        print(f'Tenemos {item} valores faltantes en {index}')

# Check to see if there are any NaN fields, use missing no to create some visualization to see these
msno.matrix(df)
plt.show()
print("----------------------------------------")
msno.bar(df)
plt.show()
msno.dendrogram(df)
plt.show()
```
